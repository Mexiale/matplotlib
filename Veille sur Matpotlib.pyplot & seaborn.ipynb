```python
Dataframes = df.copy()
X = Dataframes.drop(['coeur'], axis=1)
y = Dataframes['coeur']
```

```python
X_obj = X.select_dtypes(object)
```

```python
X_number = X.select_dtypes(exclude=object)
X_number
```

```python
encodage = LabelEncoder()
Dataframes['coeur']
```

```python
def labelEnc(x):
    if x == 'presence':
        return 1
    else:
        return 0
y = Dataframes['coeur'].map(labelEnc)
y.shape
```

```python
y.value_counts(normalize=True)*100
```

```python
#Encodage des variables qualitatives
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()

for i in X_obj:
    X_obj[i] = encodage.fit_transform(X_obj[i])
    
X_obj
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
normaliser = StandardScaler()
X_number[X_number.columns] = normaliser.fit_transform(X_number)
```

```python
X = pd.concat([X_number, X_obj], axis=1)
X
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train
```

```python
X_train
```

```python
#Création du modèle et entrainement
reg_log = LogisticRegression()
reg_log.fit(X_train,y_train)
```

```python
#Prédiction
y_pred = reg_log.predict(X_test)
y_pred
```

```python
#Evaluation (Performance du modèle)
performance = reg_log.score(X_test,y_test)
performance
```

```python
#Matrice de confusion
M_confusion = confusion_matrix(y_test, y_pred)
M_confusion
```

```python
print(classification_report(y_test, y_pred))
```

### Taux d'erreur 

```python
def taux_erreur(x1, x2):
    """
    Retourne
    """
    
    contingence = pd.crosstab(x1, x2)
    N = contingence.sum().sum()
    Num = contingence.iloc[1, 0] + contingence.iloc[0, 1]
    return round(Num/N * 100, 2)
```

```python
taux_erreur(y_test, y_pred)
```
